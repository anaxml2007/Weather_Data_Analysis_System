```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Creating a simple weather dataset
data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Temperature (Celsius)": [32, 34, 31, 35, 33, 36, 34],
}

# 2. Converting data into a clean table (DataFrame)
df = pd.DataFrame(data)
print("--- Our Weather Data ---")
print(df)
print("\n------------------------")

# 3. Plotting a beautiful line chart using the data
plt.figure(figsize=(8, 4))
plt.plot(df["Day"], df["Temperature (Celsius)"], marker="o", color="green", linewidth=2)

# Adding titles and labels in English to fix font issues
plt.title("Weekly Temperature Trends in Kerala")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.grid(True)

# Displaying the graph on screen
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Loading the dataset from weather_data.csv
df_loaded = pd.read_csv("weather_data.csv")

# 2. Setting up a professional figure size
plt.figure(figsize=(12, 5))

# 3. Plotting Humidity trends over the month using green color
plt.plot(
    df_loaded["Date"],
    df_loaded["Humidity"],
    marker="s",
    color="green",
    linewidth=2,
    label="Humidity (%)",
)

# 4. Adding professional English titles and labels
plt.title("30-Day Humidity Analysis (June 2026)")
plt.xlabel("Dates")
plt.ylabel("Humidity Percentage (%)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# 5. Displaying the green humidity graph
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 1. Loading the dataset
df_loaded = pd.read_csv("weather_data.csv")

# 2. Calculating Maximum and Average Temperature
max_temp = df_loaded["Temperature"].max()
avg_temp = df_loaded["Temperature"].mean()

# 3. Calculating Maximum and Average Humidity
max_humid = df_loaded["Humidity"].max()
avg_humid = df_loaded["Humidity"].mean()

# 4. Printing the results clearly
print("--- WEATHER DATA STATISTICS ---")
print(f"Maximum Temperature: {max_temp}°C")
print(f"Average Temperature: {avg_temp:.2f}°C")
print(f"Maximum Humidity: {max_humid}%")
print(f"Average Humidity: {avg_humid:.2f}%")
```
